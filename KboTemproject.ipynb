## Crawling Results

   ### Open Result URL

```python
kbo_game = webdriver.Chrome()
kbo_game.get("https://www.koreabaseball.com/Schedule/Schedule.aspx")
kbo_game.find_element(By.XPATH,r'//*[@id="ddlYear"]/option[2]').click()  # 연도 선택
kbo_game_html = kbo_game.page_source
result_object = BeautifulSoup(kbo_game_html, "html.parser") 
time.sleep(1)
print(result_object) 
```

### Get Data

```python
kbo_game_dict = {}
kbo_game_dict['result'] = []
d = -1

for result_link in result_object.find_all('table', attrs={'class':'tbl'}):
    # 게임 데이터 추출
    game = result_link.select('tbody > tr > td[class = "play"]')
    print("경기 데이터 추출 중...")
    for g in tqdm(range(len(game))):
        # 홈팀 및 원정 팀 데이터 추출
        Game = game[g].get_text()
        Game = Game.replace(' ','') # 경기 결과 포맷이 일정하지 않음
        team = str(re.compile('[0-9]').sub('',Game))
        team = team.split('vs')
        home = team[1]
        away = team[0]
        # 홈팀 및 원정 팀 점수 데이터 추출
        score = str(re.compile('[가-힣A-Z]').sub('',Game))  
        score = score.split('vs')
        try:
            home_score = int(score[1])
            away_score = int(score[0])
        except:
            home_score = '-'
            away_score = '-'
        
        kbo_game_dict['result'].append({'Home' : home, 'Home_Score' : home_score, 
                           'Away' : away, 'Away_Score' : away_score, 
                          'Datetime' : '', 'Stadium' : ''})
        
    
    # 경기 시간 데이터 추출
    time_list = []
    ttime = (result_link.select('tbody > tr > td[class = "time"]'))
    
    for i in range(len(ttime)):
        Time = (ttime[i].get_text())
        time_list.append(Time)
        
    # 경기장 데이터 추출    
    std = result_link.select('tbody > tr > td')
    s = -2  # 이후 경기장 정보의 인덱스가 web에서 가지는 규칙성으로 인해 어거지로 초기값 세팅
    
    # 경기 날짜 데이터 추출 후 시간 데이터와 통합해 '%Y-%m-%d %H:%M' 포맷 완성
    print('날짜데이터 & 경기장 데이터 추출중...')
    for t in tqdm(range(len(time_list))):
        if 'day' in str(result_link.select('tbody > tr')[t]):     # day 데이터가 있는 경기가 있고 없는 경기가 있어 경우를 나눠줌
            d += 1
            s += 9
            day = (result_link.select('tbody > tr > td[class="day"]'))
            Day = day[d].get_text()
            Day = str(re.compile('[가-힣()]').sub('',Day))
            Day = Day.replace('.','-')
            datetime = '2022-' + Day + ' ' + time_list[t]
            stadium = std[s].get_text()
        else:
            s += 8
            datetime = '2022-' + Day + ' ' + time_list[t]
            stadium = std[s].get_text()
        kbo_game_dict['result'][t]['Datetime'] = datetime
        kbo_game_dict['result'][t]['Stadium'] = stadium
            
    time.sleep(rd.uniform(1,2))
```

## Preprocessing

### Drop Rain Out Index

```python
# 우천취소로 인해 발생한 결측치 제거
def drop_rain_out(df):
    idx_list = []
    for raw in range(len(df)):
        for col in df.columns:
            if df.loc[raw,col] == '-':  #짝대기가있는행을 찾으면 그행을 없앰
                df = df.drop(index=raw)
                idx_list.append(raw)
                break
    df = df.reset_index(drop=True)
    return(df)
```

```python
preprocessed_kbo_game_df = drop_rain_out(kbo_game_df)
```

```python
for raw in range(len(preprocessed_kbo_game_df)):
    preprocessed_kbo_game_df.loc[raw,'Datetime'] = preprocessed_kbo_game_df.loc[raw,'Datetime'].strftime('%Y-%m-%d %H:%M:%S') #날짜 형식 변환
```

# Verifing Hypothesis

## Final DataFrame (By Merge KBO, ASOS)

```python
kbo_game_df = pd.read_csv('../data//KBO_game.csv', index_col=0)
asos_df = pd.read_csv('../data/asos.csv', index_col = 0)
```

```python
def preprocessing(df1, df2):
    """
    새로운 column 생성
    경기장 이름 -> 지역 이름
    잠실, 고척 -> 서울
    문학 -> 인천
    사직 -> 부산
    나머지는 동일
    """

    region_list = []
    for num in range(len(df1)):
        if (df1.loc[num,'Stadium'] == '잠실') or (df1.loc[num,'Stadium'] == '고척'):
            region_list.append('서울')
        elif df1.loc[num,'Stadium'] == ('문학'):
            region_list.append('인천')
        elif df1.loc[num,'Stadium'] == ('사직'):
            region_list.append('부산')
        else:
            region_list.append(df1.loc[num,'Stadium'])

    # Region Column 생성
    df1.insert(6,'Region',region_list)

    """
    야구 경기는 평균 3시간이 소요되므로,
    경기동안의 평균기온을 대치하기 위해
    14:00 경기는 15:00 기온
    17:00 경기는 18:00 기온
    18:30 경기는 20:00 기온
    """
    # 대치를 위한 Shallow Copy df 생성 후 시간 변경
    df1_copy = df1.copy()
    df1_copy['Datetime'] = df1_copy['Datetime'].str.replace('14:00','15:00')
    df1_copy['Datetime'] = df1_copy['Datetime'].str.replace('17:00','18:00')
    df1_copy['Datetime'] = df1_copy['Datetime'].str.replace('18:30','20:00')

    # 시간과 지역에 대치해 df1에 Temperature column 생성
    temperature_list = []
    for num1 in tqdm(range(df1_copy.shape[0])):
        for num2 in range(len(df2)):
            if ((df1_copy.loc[num1, 'Datetime'] == df2.loc[num2,'일시']) and (df1_copy.loc[num1, 'Region'] == df2.loc[num2,'지점명'])):
                temperature_list.append(df2.loc[num2,'기온(°C)'])

    # Temperature column 생성
    df1.insert(7,'Temperature',temperature_list)

    """
    기온과 점수간 상관관계를 구하기 위해
    Total_Score (Home_Score + Away_Score)
    column 새로 생성
    """
    total_score_list = []
    for num in range(len(df1)):
        total_score = df1.loc[num,'Home_Score'] + df1.loc[num,'Away_Score']
        total_score_list.append(total_score)

    # Total Score column 생성
    df1.insert(8,'Total_Score',total_score_list)
    
    return df1
```

```python
kbo_game_temperature_df = preprocessing(kbo_game_df, asos_df)
```

#### KBO Stat

```python
kbo_stat = webdriver.Chrome()
kbo_stat.get("https://www.koreabaseball.com/Record/Team/Hitter/Basic1.aspx")
kbo_stat.find_element(By.XPATH,r'//*[@id="cphContents_cphContents_cphContents_ddlSeason_ddlSeason"]/option[22]').click()   # 연도 선택
kbo_stat_html = kbo_stat.page_source
stat_object = BeautifulSoup(kbo_stat_html, "html.parser") 
time.sleep(1)
print(stat_object)
```

```python
kbo_stat_dict = {}
kbo_stat_dict['result'] = []

for stat_link in stat_object.find_all('table', attrs={'class':'tData tt'}):
    n = len(stat_link.select('thead > tr > th'))
    # 팀명 및 득점 데이터 추출
    data = stat_link.select('tbody > tr > td')
    for i in range(int(len(data)/n)):
        name = data[n*i+1].get_text()
        score = data[n*i+6].get_text()
        kbo_stat_dict['result'].append({'Name' : name, 'Score' : score})
        time.sleep(rd.uniform(1,2))
```

```python
print(kbo_stat_dict['result'])
```

```python
kbo_stat_json = json.dumps(kbo_stat_dict['result'], ensure_ascii = False)
kbo_stat_df = pd.read_json(kbo_stat_json)
```

```python
# 팀 당 총 득점을 경기 수로 나누어 경기당 평균 득점 column 생성
Avg_Score = []
for num in range(len(kbo_stat_df)):
    avg_score = kbo_stat_df.loc[num,'Score']/144
    Avg_Score.append(avg_score)

kbo_stat_df.insert(2,'Average_Score',Avg_Score)
```

#### Preprocess

```python
expected_score_df = kbo_game_temperature_df.copy()
```

```python
# kbo_game_temperature_df에서 홈팀, 원정팀을 kbo_stat_df와 대치시켜 경기당 총 기대 득점 계산
home_expected_score_list = []
away_expected_score_list = []
expected_score_list = []

for num1 in range(len(expected_score_df)):
    for num2 in range(len(kbo_stat_df)):
        if (expected_score_df.loc[num1, 'Home'] == kbo_stat_df.loc[num2,'Name']):
            home_expected_score = kbo_stat_df.loc[num2,'Average_Score']
            home_expected_score_list.append(home_expected_score)
        if (expected_score_df.loc[num1, 'Away'] == kbo_stat_df.loc[num2,'Name']):
            away_expected_score = kbo_stat_df.loc[num2,'Average_Score']
            away_expected_score_list.append(away_expected_score)
            
for i in range(len(home_expected_score_list)):
    expected_score = home_expected_score_list[i] + away_expected_score_list[i]
    expected_score_list.append(expected_score)
        
expected_score_df.insert(9,'Expected_Total_Score',expected_score_list)
```
